# Spotify Songs

 This dataset contains song names, artists names, link to the song and lyrics.

 ## Data

The data contains 57,650 songs with 4 columns. The columns are the artist name, the song name, a link to the lyrics, and the lyrics. In the data set there are 643 artists. There are 44,824 unique song titles. Donna Summer, Gordon Lightfoot, Bob Dylan, George Strait, and Loretta Lynn have the most songs in the data. The most common songs are *Have Yourself A Merry Little Christmas*, *Angel*, *Home*, *Hold On*, and *I Believe*.

```python
import kagglehub
import shutil
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Download latest version
path = kagglehub.dataset_download("notshrirang/spotify-million-song-dataset")
destination = r"data"
os.mkdir(destination)
for item in os.listdir(path):
    src = os.path.join(path, item)
    dest = os.path.join(destination, item)
    if os.path.isdir(src):
        shutil.copytree(src, dest, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dest)
```

```python
songs = pd.read_csv(r"data\spotify_millsongdata.csv")
songs.sample(5)
```

## Exploratory Data Analysis

```python
songs["artist"].value_counts()
```

```python
sns.countplot(
    data=songs, x="artist", order=songs["artist"].value_counts().nlargest(20).index
)
plt.xticks(rotation=90)
plt.show()
```

```python
songs["song"].value_counts()
```

```python
sns.countplot(
    data=songs, x="song", order=songs["song"].value_counts().nlargest(20).index
)
plt.xticks(rotation=90)
plt.show()
```
